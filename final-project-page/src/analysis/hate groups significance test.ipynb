```python
hh2014_count = h2014_count.rename_axis('states').reset_index(name='groups')
hh2014_count
```

```python
# 2014 hate groups:
#     The standard deviation of NY closing hours is 13.32603967176045
#     Mean = 16
#     Median = 15
#     Mode = 4, 10, 15 16
#     N = 784
#     The standard error of the mean is 0.47592998827
#     (SD = 13.32603967176045 / square root of n = 28)
#     The standard error of the sum is (SD x sq rt of n)
```

```python
_2018 = hh2018_count['groups'].tolist()
_2014 = hh2014_count['groups'].tolist()
```

```python
_2018_mean = np.mean(_2018)
_2014_mean = np.mean(_2014)
print("2018 mean value:",_2018_mean)
print("2014 mean value:",_2014_mean)
```

```python
_2018_std = np.std(_2018)
_2014_std = np.std(_2014)
print("2018 std value:",_2018_std)
print("2014 std value:",_2014_std)
```

```python
# null hypothesis is that the difference of hate groups since Trump has been elected
# isn't significant

# the alternate hypothesis is that the increase is significant
```

```python
ttest,pval = ttest_ind(_2018, _2014)
print("p-value",pval)
if pval <0.05:
  print("we reject null hypothesis")
else:
  print("we accept null hypothesis")
```
